```python
from google.colab.patches import cv2_imshow
cap = cv2.VideoCapture("/content/drive/MyDrive/test_video.mp4")

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
writer = cv2.VideoWriter('Output.mp4', cv2.VideoWriter_fourcc(*'mp4v'), 25, (width, height))

# Load model
model = YOLO("/content/drive/MyDrive/best.pt")
class_names = ['helmet', 'head']
threshold = 0.6

def get_dominant_color(image, x1, y1, x2, y2):
    """ Extracts the dominant color from the detected object. """
    roi = image[y1:y2, x1:x2]
    if roi.size == 0:
        return (0, 255, 0)  # Default to green if ROI is empty

    avg_color = np.mean(roi, axis=(0, 1))
    return tuple(int(c) for c in avg_color)  # Convert to integer BGR format

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Can't retrieve img - stream may have ended. Exiting..")
        break

    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = map(int, result[:6])
        score = math.ceil(result[4] * 100) / 100
        current_class = class_names[class_id]

        if score > threshold:
            if current_class == "helmet":
                dect_color = get_dominant_color(frame, x1, y1, x2, y2)
            else:
                dect_color = (0, 0, 255)  # Red for head (no helmet)

            cv2.rectangle(frame, (x1, y1), (x2, y2), dect_color, 3)
            cv2.putText(frame, f"{current_class} {score}", (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (255, 255, 255), 3, cv2.LINE_AA)

    writer.write(frame)
    #cv2_imshow(frame)

cap.release()
writer.release()
#cv2.destroyAllWindows()
```
